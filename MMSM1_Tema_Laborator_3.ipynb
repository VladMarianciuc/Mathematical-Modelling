###Ex. 2

Fie funcția $f(x)=x^3-7x^2+2x+4, f:[a,b]\Rightarrow\mathbb{R}$

a) Să se creeze în Python procedura **DifFinCent2**(X,Y). Parametrii de intrare sunt: vectorul $X$, reprezentând discretizarea $a = X_1 < ... < X_n = b$; vectorul $Y$, reprezentând valoarea funcției $f$ în $X$; Parametrul de ieșire fiind vectorul $d2Y$, reprezentând derivata de ordinul 2 a funcției $f$ în nodurile $x_2, x_3, ..., x_{n-1}$, calculat conform formulei de aproximare prin diferențe finite centrale pentru derivata de ordinul II.

b) Să se construiască grafic derivata de ordinul 2 a funcției $f$ pe intervalul $(-3, 7)$ și derivata de ordinul 2 obținută numeric în baza procedurii de la subpunctul a), pe intervalul $(-3, 7)$ folosind 100 de puncte.

c) Într-un alt grafic construiți eroarea, reprezentând diferența în valoarea absolută dintre derivata de ordinul 2 calculată matematic și cea calculată numeric.

```python
# Exercitiul 2
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: x**3-7*x**2+2*x+4
a=-3
b=7
x = np.linspace(a, b, 100)
h = x[1]-x[0]



fig = plt.figure()
ax = plt.axes()
ax.grid(True, color='b')

ax.plot(x, f(x), lw='3', color='r', label='y=f(x)')

def DifFinCent2(x, f, h):
  n = len(x)
  d2f = np.zeros(n)
  for i in range(n):
    d2f[i] = (f(x[i]+h)-2*f(x[i])+f(x[i]-h))/(h**2)
  return d2f

df2 = DifFinCent2(x, f, h)
ax.plot(x, df2, lw='3', color='black', label="y=f''(x)")
ax.legend(loc='best')
```

###Ex. 3

Fie următoarele date: $f(x) = sin(x),\ a=0,\ b=\pi,\ n=100$

Să se construiască grafic derivata funcției $f$ și derivata obținută numeric în baza funcției *gradient*. Într-un alt grafic construiți eroarea, reprezentând diferența în modul dintre derivata exactă și cea calculată numeric.

```python
# Exercitiul 3

import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)

a= 0
b= np.pi
x = np.linspace(a, b, 100)
h = x[1]-x[0]

fig = plt.figure()
ax = plt.axes()
ax.grid(True, color='b')

ax.plot(x, f(x), lw='3', color='r', label='y=f(x)')

def DifFinProg(x, f, h):
    n = len(x)
    df = np.zeros(n)
    for i in range(n):
        df[i]=(f(x[i]+h)-f(x[i]))/h
    return df

df = DifFinProg(x, f, h)
ax.plot(x, df, lw='3', color='g', label="y=f'(x)")

df_grad = np.gradient(f, h)
ax.plot(x, df, linestyle=':', lw='3', color='black', label="y=f'(x) gradient")
ax.legend(loc='best')
```
