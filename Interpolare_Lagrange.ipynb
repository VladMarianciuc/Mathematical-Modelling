```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sm

x = [-0.75, -0.5, -0.25, 0]
y = [-0.07181250, -0.02475000, 0.33493750, 1.101]

n = len(x)
L=[1]*n
P=0

fig = plt.figure(1)
ax = plt.axes()
x_graf=np.linspace(-0.8, 0, 100)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.grid(True)

for i in range(4):
  ax.plot(x[i], y[i], marker='o', color='b')

import sympy as sm

X = sm.Symbol('X')

for i in range(n):
  l=[1]*n
  for j in range(n):
    if(j!=i):
      l[j]=(X-x[j])/(x[i]-x[j])
    L[i]*=l[j]
  P+=y[i]*L[i]
  L[i] = sm.lambdify(X, L[i])
  ax.plot(x_graf, L[i](x_graf), label = f'L{i+1}')

P = sm.lambdify(X, P)
ax.plot(x_graf, P(x_graf), label = f'Polinomul de interpolare Lagrange')
ax.legend(loc='best')

fig2 = plt.figure(2)
ax2 = plt.axes()
x_graf=np.linspace(-0.8, 0, 100)
ax2.axhline(y=0, color='k')
ax2.axvline(x=0, color='k')
ax2.grid(True)
ax2.plot(x_graf, P(x_graf), label = f'Polinomul de interpolare Lagrange')

for i in range(n):
  ax2.plot(x[i], y[i], marker='o', color='b')
ax2.plot(-1/3, P(-1/3), marker='o', color='r', label='A(-1/3, f(-1/3))')
ax2.legend(loc='best')
print(f'Punctul de coordonata x=-1/3 este interpolat la coordonata y={P(-1/3)}')
```
